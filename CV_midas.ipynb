### Import PyTorch related modules

```
%pylab inline
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.utils.data.dataloader as dataloader
import torch.optim as optim

from torch.utils.data import TensorDataset
from torch.autograd import Variable
from torchvision import transforms
```

## CUDA setup 
### Seed for reproducibilty

```
SEED = 1

# CUDA?
cuda = torch.cuda.is_available()

# For reproducibility
torch.manual_seed(SEED)

if cuda:
    torch.cuda.manual_seed(SEED)
```

**CNN architecture**

```
class ConvNet(nn.Module):
    def __init__(self):
        super(ConvNet, self).__init__()
        self.conv1 = nn.Conv2d(1, 20, 5, 1)  # 20 filters, kernel size = 5*5
        self.conv2 = nn.Conv2d(20, 50, 5, 1) # 50 filters, kernel size = 5*5
        self.fc1 = nn.Linear(4*4*50, 500)    # Map filters to a linear layer (MLP)
        self.fc2 = nn.Linear(500, 4) 

    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = F.max_pool2d(x, 2, 2)
        x = F.relu(self.conv2(x))
        x = F.max_pool2d(x, 2, 2)
        x = x.view(-1, 4*4*50)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x
    
    def name(self):
        return "ConvNet"
```

**Initialize model and optimizer**

```
model = ConvNet()
if cuda:
    model.cuda() # CUDA!
optimizer = optim.Adam(model.parameters(), lr=1e-3)
```

### Dataset class for Torch DatasetLoader (Batch Generation)

```
import torch
from torch.utils import data

class Dataset(data.Dataset):
  'Characterizes a dataset for PyTorch'
  def __init__(self,train_data,list_IDs, labels):
        'Initialization'
        self.labels = labels
        self.list_IDs = list_IDs

  def __len__(self):
        'Denotes the total number of samples'
        return len(self.list_IDs)

  def __getitem__(self, index):
        'Generates one sample of data'
        # Select sample
        ID = self.list_IDs[index]

        # Load data and get label
        X = torch.Tensor(train_data[ID])
        y = self.labels[ID]

        return X, y
```

**Split training dataset into train and validation**

```
partition = dict()

train_ratio = 0.9
train_size = int(train_ratio*train_data.shape[0])     # 0.9 * 8000

partition['train'] = list(range(0,train_size))        # Store list of indexes for train and validation for batch generator
partition['validation'] = list(range(train_size,train_data.shape[0]))
```

```
training_set = Dataset(train_data,partition['train'], train_labels) 
validation_set = Dataset(train_data,partition['validation'], train_labels)
```

```
# Create DataLoader
dataloader_args = dict(shuffle=True, batch_size=64,num_workers=4, pin_memory=True) if cuda else dict(shuffle=True, batch_size=64)
train_loader = dataloader.DataLoader(training_set, **dataloader_args)
val_loader = dataloader.DataLoader(validation_set, **dataloader_args)
```

```
EPOCHS = 20
losses = []



model.train()
for epoch in range(EPOCHS):
    for batch_idx, (data, target) in enumerate(train_loader):
        # Get Samples
        data, target = Variable(data), Variable(target)
        
        if cuda:
            data, target = data.cuda(), target.cuda()
        
        # Init
        optimizer.zero_grad()

        # Predict
        y_pred = model(data) 

        # Calculate loss
        loss = F.cross_entropy(y_pred, target)
        losses.append(loss.cpu().item())
        # Backpropagation
        loss.backward()
        optimizer.step()
        
        
        # Display
        if batch_idx % 100 == 1:
            print('\r Train Epoch: {}/{} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                epoch+1,
                EPOCHS,
                batch_idx * len(data), 
                len(train_loader.dataset),
                100. * batch_idx / len(train_loader), 
                loss.cpu().item()), 
                end='')
            
      
      
    # Eval
    with torch.set_grad_enabled(False):
      tot_num = 0
      correct_cnt = 0
      
      for batch_idx, (data, target) in enumerate(val_loader):
          # Get Samples
          data, target = Variable(data), Variable(target)

          if cuda:
              data, target = data.cuda(), target.cuda()

          # Init
          optimizer.zero_grad()

          # Predict
          y_pred = model(data) 
          
          _, pred_label = torch.max(y_pred.data, 1)
          

          correct_cnt += (pred_label == target.data).sum()
          tot_num += pred_label.size()[0]

          
      accuracy = float(correct_cnt.cpu().numpy() / tot_num)

      
      
      
    
    print('\r Train Epoch: {}/{} \tLoss: {:.6f}\t Val Accuracy: {:.4f}%'.format(
        epoch+1,
        EPOCHS,
        loss.cpu().item(),
        accuracy*100,
        end=''))
```
